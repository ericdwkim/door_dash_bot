```python
import re
import pandas as pd
import numpy as np

def parse_store_info_list_to_df(df):
    store_num = []
    store_hint = []
    pattern = re.compile(r"[TТ][XХ][BВ]\s*#(\d+)(?:\s*\((.*?)\)|\s*(.*))?", re.IGNORECASE)

    for idx, row in df.iterrows():
        match = re.search(pattern, row['store_info'])

        if match:
            store_num.append(int(match.group(1)))
            if match.group(2):
                store_hint.append(match.group(2))
            else:
                store_hint.append(match.group(3))
        else:
            store_num.append(np.nan)
            store_hint.append(np.nan)

    df['store_num'] = store_num
    df['store_hint'] = store_hint
    
    df['store_num'] = df['store_num'].astype('Int64')  # Use Int64 to support NaN
    df['store_hint'] = df['store_hint'].astype(str)

    return df

store_info_list_df = parse_store_info_list_to_df(store_info_list_df_raw)
```
